```python
#Applying Ml Algorithms
```

```python
X = data_cl.drop("Has_Hypertension", axis=1) 
y = data_cl["Has_Hypertension"] 
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=50)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
alg = LogisticRegression()
alg.fit(X_train, y_train)
pred_alg=alg.predict(X_test)
```

```python
print("Logistic Regression Accuracy:", accuracy_score(y_test, pred_alg)*100,'%')
```

```python
#applying decision tree
from sklearn.tree import DecisionTreeClassifier
tree = DecisionTreeClassifier()
tree.fit(X_train, y_train)
pred_tree = tree.predict(X_test)
print("Decision Tree Accuracy:", accuracy_score(y_test, pred_tree)*100,'%')
```

```python
#applying Randomforest
from sklearn.ensemble import RandomForestClassifier
rdf = RandomForestClassifier()
rdf.fit(X_train, y_train)
pred_rdf = rdf.predict(X_test)
print("Random Forest Accuracy:", accuracy_score(y_test, pred_rdf)*100,'%')
```

```python
#applying Support vector mechine
from sklearn.svm import SVC
svma = SVC()
svma.fit(X_train, y_train)
pred_svma = svma.predict(X_test)
print("SVM Accuracy:", accuracy_score(y_test, pred_svma)*100,'%')
```

```python
from sklearn.metrics import classification_report
print("Logistic Regression Report:\n", classification_report(y_test,pred_alg))
print("Decision Tree Report:\n", classification_report(y_test, pred_tree))
print("Random Forest Report:\n", classification_report(y_test, pred_rdf))# this is best  accuracy
print("SVM Report:\n", classification_report(y_test, pred_svma))
```
